```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
prompt = "A futuristic smart city at night with glowing buildings, flying cars, neon lights, ultra realistic, 4K"

image = pipe(prompt).images[0]
image
```

```python
prompt = "A futuristic AI-powered hospital with advanced medical robots, holographic health screens, doctors using augmented reality, ultra realistic, 4K, cinematic lighting"
image = pipe(prompt).images[0]
image
```
